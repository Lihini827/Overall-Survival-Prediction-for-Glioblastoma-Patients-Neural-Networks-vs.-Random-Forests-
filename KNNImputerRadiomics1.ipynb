```python
import pandas as pd
import numpy as np
```

```python
path='/content/drive/MyDrive/gbm/ImpPreProcess_data.csv'
```

```python
df=pd.read_csv(path)
df.head()
```

```python
print(df.columns)
```

```python
from sklearn.impute import KNNImputer

imp = KNNImputer(missing_values=np.nan,
                 n_neighbors=5,
                 weights='uniform',
                 metric='nan_euclidean',
                 copy=True,
                 add_indicator=False,
                 keep_empty_features=False)
imp.fit(df)
```

```python
import pandas as pd
```

```python
print(type(df))
```

```python
if type(df) != pd.DataFrame:
    df = pd.DataFrame(df)
```

```python
numerical = df.columns
```

```python
df=imp.transform(df)
df=pd.DataFrame(df,columns=numerical)
df.head()
```
